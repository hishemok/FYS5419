```python
from qiskit import QuantumCircuit, transpile, assemble
from qiskit.visualization import circuit_drawer
import qiskit_aer
import matplotlib.pyplot as plt

bell = QuantumCircuit(2, 2)
bell.h(0)
bell.cx(0, 1)
bell.measure([0, 1], [0, 1])
print("Quantum Circuit")
print(bell)

num_shots = 10000
simulator = qiskit_aer.Aer.get_backend("qasm_simulator")
transpiled_circuit = transpile(bell, simulator)
result = simulator.run(transpiled_circuit,shots=num_shots).result()
counts = result.get_counts(bell)

print(counts)
plt.bar(counts.keys(), counts.values())
plt.title("Bell state measurement results")
plt.ylabel("Counts")
plt.show()
```
